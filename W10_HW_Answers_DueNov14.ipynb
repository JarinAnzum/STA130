### Question 1

1. Simple Linear Regression models a linear relationship between a single predictor variable $x$ and single outcome variable $y$.

$$\text{Outcome} = \beta_0 + \beta_1 \times \text{Predictor}$$

Multiple Linear Regression models a linear relationship between multiple predictor variables (for example $x_1$, $x_2$, $x_3$) and a single outcome variable $y$. 

$$\text{Outcome} = \beta_0 + \beta_1 \times \text{Predictor}_1 + \beta_2 \times \text{Predictor}_2 + \dots + \beta_n \times \text{Predictor}_n
$$

The benefit of Muliple Linear Regression over Simple Linear Regression is that it is able to represent complex relationships in data, leading to more well-informed improved prediction. 

2. In Simple Linear Regression, a continuous variable is a variable that can take on any value within a given range (a variable that represents a continuous value). An indicator variable on the other hand is a variable that represents qualitative categorical data. Specifically, it represents binary categorical data where 0 can indicate the absence of a characteristic, and 1 can indicate the presence.

3. Simple Linear Regression is based on a single continuous predictor variable. The linear calculation represents the linear relationship between this continuous predictor and an outcome.

Here is an example of what this may look like: 

$$\text{Salary} = \beta_0 + \beta_1 \times \text{Experience}$$

Multiple Linear Regression is based on an indicator variable that splits the model into 2 groups (given that the indicator is usually binary). Each group has its own intercept, but they share the same slope.

Here is an example of what this may look like:

$$\text{Salary} = \beta_0 + \beta_1 \times \text{Experience} + \beta_2 \times 1(\text{Male})$$

So for women, the line is...

$$\text{Salary} = \beta_0 + \beta_1 \times \text{Experience}
$$

...and for men, the line is...

$$\text{Salary} = (\beta_0 + \beta_2) + \beta_1 \times \text{Experience}$$

Contextually, this can mean that men and women start at different points (different intercepts), but the way experience affects salary is the same for both groups (same slope). 

4. When an interaction term is added to a regression model, we are changing the relationship (slope) between the predictor variable and outcome to change, depending on the category.

Relating to the example above, we are not only changing the intercept, we are also slope between groups.

$$\text{Outcome} = \beta_0 + \beta_1 \times \text{Continuous Predictor} + \beta_2 \times 1(\text{Indicator}) + \beta_3 \times (\text{Continuous Predictor} \times 1(\text{Indicator}))
$$

Here, the interaction term $\beta_3$ lets the slope be different for each group. This means that the line of each group are no longer parallel, and instead, are intersecting lines.

Contextually, we can assume that men and women not only start at different points (different intercepts), but the way experience affects salry is different for each group (different slopes).

5. When we want to use non-binary categorical variables (like region: North, South, East, West) in a regression model, we can’t directly plug them in as words. Instead, we create indicator variables (also called dummy variables) to represent each category as 0s and 1s.

- If there are $k$ categorical variables, we need $k-1$ indicator variables.
- A single category is left out as baseline group, which is the line other lines are compared to.

##### Same Slope

Here is what this looks like if all the groups have the same slope...

$$\text{Outcome} = \beta_0 + \beta_1 \times 1(\text{South}) + \beta_2 \times 1(\text{East}) + \beta_3 \times 1(\text{West})
$$

$\beta_0$ represents the baseline group's intercept. The other $\beta$'s are how each group differs from the baseline.

##### Different Slopes

...and here is what this looks like if all the groups have different slopes...

$$\text{Outcome} = \beta_0 + \beta_1 \times \text{Continuous Predictor} + \beta_2 \times 1(\text{Category}_1) + \beta_3 \times 1(\text{Category}_2) + \dots + \beta_{k-1} \times 1(\text{Category}_{k-1}) + \gamma_1 \times (\text{Continuous Predictor} \times 1(\text{Category}_1)) + \gamma_2 \times (\text{Continuous Predictor} \times 1(\text{Category}_2)) + \dots
$$

Note that the $\gamma_1, \gamma_2$ are the differences in slopes when compared to the baseline. They are not different slopes themselves!

Some Additional Information:

- We use $k-1$ indicator variables to avoid the "dummy variable trap" wherein including $k$ indicators leads to perfect multicollinearity (redundant information that breaks the model).

The model will examine:
- The individual effects of `Sp. Def` and `Generation`.
- Interaction effects between `Sp. Def` and `Generation`.

```python
import statsmodels.formula.api as smf

# Specify and fit the model
model2_spec = smf.ols(formula='HP ~ Q("Sp. Def") * C(Generation)', data=pokeaman)
model2_fit = model2_spec.fit()

# Display the summary
model2_fit.summary()
```

We know that... $R^2$ and coefficient significance assess different aspects of a model. $R^2$ represents the proportion of variance explained by the model while coefficients' p-values signify the strength of relationships between predictors and outcome (with p-value being the likelihood of getting value as extreme if the null hypothesis were true).

For our Pokemon model, here is what these values represent specifically:

- Coefficients describe the average change in `HP` for a unit change in the predictor (holding others constant).

- Small p-values indicate significant evidence against the null hypothesis (no effect).

- $R^2$ describes the proportion of variability in `HP` explained by the model (with low $R^2$ suggesting that the model doesn't explain much of the total variation in `HP`).
